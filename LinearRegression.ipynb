```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
sns.set()
```

```python
data = pd.read_csv('files/Linear Regression - Sheet1.csv')
```

```python
data.head()
```

```python
plt.plot(data.X,data.Y)
plt.show()
```

```python
ls = LinearRegression()
x_train = data['X'].values.reshape(-1,1)
y_train = data['Y'].values.reshape(-1,1)
```

```python
ls_model = ls.fit(x_train,y_train)
```

```python
print('Formula for Linear Regression is y = mx + c')
c = ls.intercept_[0]
m = ls.coef_[0][0]
print('c = ' + str(c))
print('m = ' + str(m))
```

```python
y_predict = ls.predict(x_train)
```

```python
from sklearn.metrics import r2_score
print('r2 Score: ' + str(r2_score(y_train,y_predict)))
```

```python
plt.plot(x_train,y_train,color='red')
plt.plot(x_train,y_predict,color='green')
plt.show()
```
